```python
filepath = '/content/Countries.csv'
df = pd.read_csv(filepath)
print(df)
```

```python
# Display first 5 rows
print("First 5 Rows")
print(df.head())

# Display last 5 rows
print("\nLast 5 Rows")
print(df.tail())

# Shape of dataset
print("\nShape of Dataset")
print(df.shape)

# Column names
print("\nColumn Names")
print(df.columns)

# Dataset Information
print("\nDataset Information")
print(df.info())

# Summary Statistics
print("\nSummary Statistics")
print(df.describe())

# Data Types
print("\nData Types")
print(df.dtypes)

# Missing Values
print("\nMissing Values")
print(df.isnull().sum())

# Number of Unique Values
print("\nUnique Values")
print(df.nunique())

# Random Sample
print("\nRandom Sample")
print(df.sample(5))

# Value Counts
print("\nValue Counts")
print(df["country"].value_counts())
```
